Ассиметрия положительная, что указывает на то, что хвост находится на правой стороне распределения, которая простирается в сторону более положительных значений. Данное распределение имеет тенденцию производить больше выбросов, чем нормальное распределение. Это объясняется большим объемом выборки.

```python
percentages = [0.05, 0.1, 0.25, 0.5, 0.75, 0.8, 0.95]
percentiles_df = pd.DataFrame(df_google[['Minimum Installs', 'Maximum Installs']].quantile(percentages))
percentiles_df.index = [f'{p * 100}%' for p in percentages]
print(percentiles_df.transpose())
```

Наибольшее количество значений у мин. кол-ва скачиваний лежит в интервале от 50 до 5000, у макс. кол-ва скачиваний - в интервале от 84 до 7353.

```python
import matplotlib.pyplot as plt
```

```python
ax = (df_google['Minimum Installs']).plot.box(figsize=(13, 14))
ax.set_ylabel('Minimum Installs')

plt.tight_layout()
plt.show()
```

```python
ax = (percentiles_df['Minimum Installs']/1000).plot.box(figsize=(13, 14))
ax.set_ylabel('Minimum Installs')

plt.tight_layout()
plt.show()
```

```python
ax = (df_google['Maximum Installs']).plot.box(figsize=(13, 14))
ax.set_ylabel('Maximum Installs')

plt.tight_layout()
plt.show()
```

Мы видим, что тут значительно большее количество выборосов, нежели в минимальном количестве скачиваний.

```python
ax = (percentiles_df['Maximum Installs']/1000).plot.box(figsize=(13, 14))
ax.set_ylabel('Maximum Installs')

plt.tight_layout()
plt.show()
```
